```python
from pathlib import Path
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GroupKFold, GridSearchCV
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
import shap
import matplotlib.pyplot as plt
import seaborn as sns
from yahooquery import search
from tqdm import tqdm
```

```python
ROOT = Path().resolve().parent    # go up one level from src/
RAW = ROOT / "data" / "raw"
PROCESSED = ROOT / "data" / "processed" 

DATA_MODEL = pd.read_parquet(PROCESSED / "FULL_MODEL_INPUTS_OUTPUT.parquet")
DATA_MODEL
```

# Random Forest

```python
# DataFrame for Sample 1A
model_df = DATA_MODEL.copy()

model_df = model_df[model_df["DQS"] > 0]

model_df["PRSxDQS"] = model_df["PRS"] * model_df["DQS"]

target_col = "sigma_idio_ann"

feature_cols = [
    "PRS",
    "DQS",
    "PRSxDQS",
    "leverage",
    "equity_ratio",
    "asset_growth",
    "ROA",
    "ROE",
    "book-to-market",
    "cash_ratio",
    "log_mktcap",
]

# Drop rows with missing target or any missing feature (simple baseline)
model_df = model_df.dropna(subset=[target_col] + feature_cols).copy()

# --- RANDOM FOREST ---

# 1) Prepare Data (Sorted Chronologically) for Random Forest 
# Ensure your index is reset and sorted
model_df = model_df.sort_values(by=["year", "PERMNO"]).reset_index(drop=True)

firm_number = len(model_df.PERMNO.unique())
print(f"Total number of firms with climate disclosure in the model (with valid data) : {firm_number}")

# Define Cutoff for Final Test (2023-2024)
holdout_mask = model_df["year"] >= 2023
dev_mask     = model_df["year"] < 2023

X_dev = model_df.loc[dev_mask, feature_cols]
y_dev = model_df.loc[dev_mask, target_col]

X_test = model_df.loc[holdout_mask, feature_cols]
y_test = model_df.loc[holdout_mask, target_col]

print(f"Development Set (Train+Tune): {X_dev.shape}")
print(f"Test Set (Final Eval):        {X_test.shape}")
```

```python
# --- HYPERPARAMETER TUNING ---

# 2. Define TimeSeriesSplit (Walk-Forward)
tscv = TimeSeriesSplit(n_splits=5)

# 3. Define the Model
rf = RandomForestRegressor(random_state=42)

# 4. Define Hyperparameter Grid
# Focused grid to prevent overfitting on financial noise
param_grid = {
    'n_estimators': [200, 500],        # Sufficient trees to stabilize variance
    'max_depth': [4, 6, 8],            # Shallow trees to avoid memorization
    'min_samples_leaf': [10, 20, 50],  # Force robustness (don't fit outliers)
    'max_features': ['sqrt', 'log2']   # Randomness in split selection
}

# 5. Grid Search
rf_grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=tscv,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)

print("Starting Walk-Forward Grid Search...")
rf_grid.fit(X_dev, y_dev)

print("\nBest Parameters found:")
print(rf_grid.best_params_)
```

```python
# --- MODEL EVALUATION (Benchmark: ElasticNet R2 = 18.3%) ---

# 6. Evaluate on Unseen Test Set (2023-2024)
best_rf = rf_grid.best_estimator_
y_pred_test = best_rf.predict(X_test)

r2_test = r2_score(y_test, y_pred_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

print("-" * 30)
print(f"Random Forest Test R^2: {r2_test:.4f}")
print(f"Random Forest Test RMSE: {rmse_test:.4f}")
print("-" * 30)
```

```python
# 7. Feature Importance (Gini) -> tells  "how often" a feature was used to make a decision, but not "how" (positive vs. negative).

importances = best_rf.feature_importances_
feat_imp = pd.DataFrame({
    'feature': feature_cols,
    'importance': importances
}).sort_values('importance', ascending=False)

print("\nGini Feature Importance:")
print(feat_imp)
```

```python
# 8. SHAP Analysis 
# -> reveal the direction and interaction effects (Does high DQS lower volatility? Does DQS affect high-risk companies differently?)
# Initialize Explainer (TreeExplainer is optimized for Random Forests)
explainer = shap.TreeExplainer(best_rf)

# Calculate SHAP values for the TEST set
# (We want to see how the model treats unseen data)
shap_values = explainer.shap_values(X_test)

# A. Summary Plot (Global Importance + Direction)
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, show=False)
plt.title("SHAP Summary: Which Features Drive Volatility?")
plt.show()

# B. Dependence Plot: DQS vs Volatility
# Does the relationship look U-shaped? Is it flat?
plt.figure(figsize=(8, 5))
shap.dependence_plot("DQS", shap_values, X_test, show=False)
plt.title("Dependence Plot: Effect of DQS on Volatility")
plt.show()

# C. Interaction Plot: PRS vs DQS
# This is the money plot: How does Disclosure modify the market's reaction to Physical Risk?
# We plot SHAP values of PRS, colored by DQS.
plt.figure(figsize=(8, 5))
shap.dependence_plot(
    "PRS", 
    shap_values, 
    X_test, 
    interaction_index="DQS",
    show=False
)
plt.title("Interaction: Impact of Physical Risk (PRS) colored by Disclosure Quality (DQS)")
plt.show()
```

# Gradient Boosting Regressor

```python

# ---------------------------------------------------------
# 1. Setup & Data (Reuse existing splits)
# ---------------------------------------------------------

print(f"Training on Dev Set: {X_dev.shape}")
print(f"Testing on Holdout:  {X_test.shape}")

# ---------------------------------------------------------
# 2. Define Model & Hyperparameters
# ---------------------------------------------------------
# GBR requires balancing 'learning_rate' with 'n_estimators'.
# Lower learning_rate (0.01) needs more trees (500+) but generalizes better.

gbr = GradientBoostingRegressor(random_state=42)

param_grid = {
    'n_estimators': [200, 500],       # Boosting needs more iterations than RF
    'learning_rate': [0.01, 0.05, 0.1], # The "step size" of correction
    'max_depth': [3, 4, 5],           # Keep distinctively shallow (Boosting overfits deep trees)
    'subsample': [0.8, 1.0],          # Stochastic boosting (0.8) often robustifies against noise
    'min_samples_leaf': [10, 20]      # Regularization constraint
}

# ---------------------------------------------------------
# 3. Walk-Forward Grid Search
# ---------------------------------------------------------
# We use the EXACT same TimeSeriesSplit as the Random Forest for fair comparison
tscv = TimeSeriesSplit(n_splits=5)

gbr_grid = GridSearchCV(
    estimator=gbr,
    param_grid=param_grid,
    cv=tscv,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)

print("Starting Gradient Boosting Walk-Forward Search...")
gbr_grid.fit(X_dev, y_dev)

print("\nBest GBR Parameters:")
print(gbr_grid.best_params_)

# ---------------------------------------------------------
# 4. Final Evaluation (Test Set)
# ---------------------------------------------------------
best_gbr = gbr_grid.best_estimator_
y_pred_test_gbr = best_gbr.predict(X_test)

r2_test_gbr = r2_score(y_test, y_pred_test_gbr)
rmse_test_gbr = np.sqrt(mean_squared_error(y_test, y_pred_test_gbr))

print("-" * 40)
print(f"Gradient Boosting Test R^2: {r2_test_gbr:.4f}")
print(f"Gradient Boosting Test RMSE: {rmse_test_gbr:.4f}")
print("-" * 40)

# # ---------------------------------------------------------
# # 5. SHAP Interpretability
# # ---------------------------------------------------------
# # Gradient Boosting is also tree-based, so we use TreeExplainer.
# explainer_gbr = shap.TreeExplainer(best_gbr)
# shap_values_gbr = explainer_gbr.shap_values(X_test)

# # A. Summary Plot (Compare feature ranking vs RF)
# plt.figure(figsize=(10, 6))
# shap.summary_plot(shap_values_gbr, X_test, show=False)
# plt.title("GBR: SHAP Summary Plot")
# plt.show()

# # B. DQS Non-Linearity (Compare the 'Smile' shape)
# plt.figure(figsize=(8, 5))
# shap.dependence_plot("DQS", shap_values_gbr, X_test, show=False)
# plt.title("GBR Dependence: DQS vs Volatility")
# plt.show()

# # C. Interaction (PRS x DQS)
# plt.figure(figsize=(8, 5))
# shap.dependence_plot(
#     "PRS", 
#     shap_values_gbr, 
#     X_test, 
#     interaction_index="DQS",
#     show=False
# )
# plt.title("GBR Interaction: Physical Risk (PRS) colored by DQS")
# plt.show()
```

```python
# ---------------------------------------------------------
# 5. SHAP Interpretability
# ---------------------------------------------------------
# Gradient Boosting is also tree-based, so we use TreeExplainer.
explainer_gbr = shap.TreeExplainer(best_gbr)
shap_values_gbr = explainer_gbr.shap_values(X_test)

# A. Summary Plot (Compare feature ranking vs RF)
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values_gbr, X_test, show=False)
plt.title("GBR: SHAP Summary Plot")
plt.show()

# B. DQS Non-Linearity (Compare the 'Smile' shape)
plt.figure(figsize=(8, 5))
shap.dependence_plot("DQS", shap_values_gbr, X_test, show=False)
plt.title("GBR Dependence: DQS vs Volatility")
plt.show()

# C. Interaction (PRS x DQS)
plt.figure(figsize=(8, 5))
shap.dependence_plot(
    "PRS", 
    shap_values_gbr, 
    X_test, 
    interaction_index="DQS",
    show=False
)
plt.title("GBR Interaction: Physical Risk (PRS) colored by DQS")
plt.show()
```

# Portfolio Application

```python
df_2024_filings = DATA_MODEL.copy()

df_2024_filings = df_2024_filings[df_2024_filings["DQS"] > 0]

df_2024_filings = df_2024_filings[df_2024_filings["year"] == 2024]

df_2024_filings["PRSxDQS"] = df_2024_filings["PRS"] * df_2024_filings["DQS"]

# Drop rows with missing target or any missing feature (simple baseline)
df_2024_filings = df_2024_filings.dropna(subset=[target_col] + feature_cols).copy()

# --- RANDOM FOREST ---

# 1) Prepare Data (Sorted Chronologically) for Random Forest 
# Ensure your index is reset and sorted
df_2024_filings = df_2024_filings.sort_values(by=["year", "PERMNO"]).reset_index(drop=True)

firm_number = len(df_2024_filings.PERMNO.unique())
print(f"Total number of firms with climate disclosure in the model (with valid data) : {firm_number}")

df_2024_filings
```

```python
# 1. TRAIN the model on History (Strictly < 2024)
# We use X_dev (2016-2022) or you can retrain on everything up to 2023.
# Let's assume 'best_gbr' is trained on 2016-2023 data mapping to T+1 volatility.

# 2. LOAD the "Live" Data (2024 Filings)
# This represents the information available to you in early 2025.
# X_2024 contains:
#   - PRS/DQS extracted from the 2024 10-K (filed Feb 2025)
#   - Financials (ROA, Size) from year-end 2024

X_2024 = df_2024_filings[feature_cols].copy()

# 3. PREDICT 2025 Risk (The "Smart" Signal)
# The model has never seen 2024 data or 2025 volatility.
predicted_2025_vol = best_gbr.predict(X_2024)

# 4. CONSTRUCT the Portfolio (Quintile Sort)
# Rank companies by how risky the model THINKS they will be in 2025.
portfolio_df = df_2024_filings.copy()
portfolio_df['Signal_Vol'] = predicted_2025_vol

# Create 5 Buckets (0=Safest Prediction, 4=Riskiest Prediction)
portfolio_df['Bucket'] = pd.qcut(portfolio_df['Signal_Vol'], 5, labels=False)

# 5. DEFINE LEGS
# LONG: The "Climate Bunkers"
# Companies with LOW Physical Risk OR Moderate Risk + "Safe" Disclosure
long_leg = portfolio_df[portfolio_df['Bucket'] == 0]

# SHORT: The "Revelation Bombs"
# Companies with HIGH Physical Risk + HIGH DQS (The Amplifier Effect)
short_leg = portfolio_df[portfolio_df['Bucket'] == 4]

print(f"Strategy Ready for 2025 Trading:")
print(f"Long Positions: {len(long_leg)} companies")
print(f"Short Positions: {len(short_leg)} companies")
```

```python
mktcap_2024 = pd.read_csv(PROCESSED / "mktcap_2024.csv")
mktcap_2024.index = pd.to_datetime(mktcap_2024.date)
mktcap_2024.index.name = "date"
mktcap_2024 = mktcap_2024.drop(columns="date")

# convert to millions
mktcap_fixed = mktcap_2024.div(1000)

# keep last row only
mktcap_fixed = mktcap_fixed.iloc[-1:]
```

```python
mktcap_fixed.columns = mktcap_fixed.columns.astype(int)

# keep only long & short legs in the mktcap dataframe
LONG_MASK = LONG_LEG["PERMNO"].astype(int).tolist()
LONG_MKTCAP = mktcap_fixed.loc[:, mktcap_fixed.columns.intersection(LONG_MASK)]

SHORT_MASK = SHORT_LEG["PERMNO"].astype(int).tolist()
SHORT_MKTCAP = mktcap_fixed.loc[:, mktcap_fixed.columns.intersection(SHORT_MASK)]

# Long leg
long_total = LONG_MKTCAP.sum(axis=1)        # sum across columns (per row)
LONG_WEIGHTS = LONG_MKTCAP.div(long_total, axis=0)
LONG_WEIGHTS = LONG_WEIGHTS.squeeze()
print(LONG_WEIGHTS.sum())

# Short leg
short_total = SHORT_MKTCAP.sum(axis=1)
SHORT_WEIGHTS = SHORT_MKTCAP.div(short_total, axis=0)
SHORT_WEIGHTS = SHORT_WEIGHTS.squeeze()
print(SHORT_WEIGHTS.sum())

# turn into balanced long-short
#LONG_WEIGHTS  = LONG_WEIGHTS * 0.5
#SHORT_WEIGHTS = SHORT_WEIGHTS * -0.5

#WEIGHTS = pd.concat([LONG_WEIGHTS, SHORT_WEIGHTS], axis=0)
#WEIGHTS.name = "weight"

#WEIGHTS = WEIGHTS.squeeze()
#print(WEIGHTS.sum())

WEIGHTS = LONG_WEIGHTS
```

```python
# 1. Clean returns as you already do
RETURNS_2025 = prices_fixed.pct_change()
RETURNS_2025.replace([np.inf, -np.inf], np.nan, inplace=True)
RETURNS_2025 = RETURNS_2025.fillna(0)
RETURNS_2025.columns = RETURNS_2025.columns.astype(int)

# 2. Align to initial weights (missing columns get 0 returns)
RETURNS_2025_ALIGNED = RETURNS_2025.reindex(columns=WEIGHTS.index, fill_value=0)

# 3. Build cumulative gross returns for each asset
gross = 1 + RETURNS_2025_ALIGNED
cum_gross = gross.cumprod()

# 4. Start with initial weights WEIGHTS (long positive, short negative)
#    Portfolio value at time t = sum_i w_i * cum_gross_{i,t}
PORTFOLIO_VALUE_2025 = (cum_gross * WEIGHTS.values).sum(axis=1)

# Optional: normalize to 1 at start
PORTFOLIO_VALUE_2025 = PORTFOLIO_VALUE_2025 / PORTFOLIO_VALUE_2025.iloc[0]

# 5. Derive portfolio returns from that value series
PORTFOLIO_RET_2025 = PORTFOLIO_VALUE_2025.pct_change().fillna(0)

PORTFOLIO_VALUE_2025.plot()
```
